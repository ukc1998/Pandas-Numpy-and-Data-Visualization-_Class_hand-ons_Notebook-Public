```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(20)
df = pd.DataFrame(np.random.randn(10,4), index=pd.date_range("2022-12-10", periods=10), columns=["a", "b", "c", "d"] )
df
```

```python
df["new"] = df.c + df.d
```

```python
df
```

```python
df["square_d"] = df["d"].apply(lambda x: x**2)
```

```python
df
```

```python
df["log_d"] = df["d"].apply(lambda x: np.log(x))
df
```

```python
def test(x):
    if x<0:
        return x
    else:
        return np.log(x)
```

```python
df["log_d"] = df["d"].apply(lambda x: np.log(x))
df
```

```python
df["log_d_1"] = df["d"].apply(test)
```

```python
df
```

```python
df.dtypes
```

```python
def test1(x):
    if x < 0:
        return 1
    else:
        return 0
```

```python
df["new_n"] = df["a"].apply(test1)
```
