```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.offline import iplot
import plotly as py
import plotly.tools as tls
import cufflinks as cf
%matplotlib inline
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.load_dataset("tips")
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.relplot(x = "tip", y = "total_bill", data = tips)
```

```python
# grouping based whether if smoker or not 
sns.relplot(x = "tip", y = "total_bill", data = tips, hue = "smoker")
```

```python
# grouping based on day
sns.relplot(x = "tip", y = "total_bill", data = tips, hue = "day")
```

```python
# grouping based on sex
sns.relplot(x = "tip", y = "total_bill", data = tips, hue = "sex")
```

```python
# grouping based on two factors: smoker  plus sex
sns.relplot(x = "tip", y = "total_bill", data = tips, hue = "smoker", style="sex")
```

```python
# grouping based on two factors: smoker  plus day
sns.relplot(x = "tip", y = "total_bill", data = tips, hue = "day", style="smoker")
```

```python
# more factors takenn into account
sns.relplot(x = "tip", y = "total_bill", data = tips, hue = "day", style="smoker", size = "tip")
```
